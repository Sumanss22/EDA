### Name the top 8 industries with maximum count of billionaires

```python
df.head(1)
```

```python
df_Ind = df['Industries'].value_counts().head(8)
df_Ind
```

```python
#Alternate
top_8 = df.groupby('Industries').nunique().sort_values('Name',ascending=False).reset_index().head(8)
top_8
```

### Compare the mean of net worth of billionaires from these top 8 industries.

```python
#This is for all industries and not Top 8 found in last question
df[['Industries','Net Worth']].groupby('Industries').mean().sort_values('Net Worth',ascending=False).head(8)
```

```python
df_new = pd.DataFrame(columns=df.columns)
df_new
```
